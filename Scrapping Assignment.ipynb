```python
import pandas as pd
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.cm as cm
import matplotlib.colors as colors

import folium
from geopy.geocoders import Nominatim


dataframe = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')

df = dataframe[0]
df
```


<H2> Removing Values that are "Not Assigned" from the column Borough

```python
df.set_index('Borough', inplace = True)
dff= df.drop('Not assigned',axis = 0)
dff.head(10)
```

<H2>Grouping Values in Neighbourhood column seperated by commas

```python
ds = dff.groupby(['Postcode','Borough'], as_index=True).agg(','.join)
ds.reset_index(inplace = True) 
ds
```

```python

ds.loc[ds.Neighbourhood == 'Not assigned', 'Neighbourhood'] = ds['Borough']
ds1= ds.rename(columns={'Postcode':'Postal Code'})
ds1
```

<H2> Using .shape method to find out number of rows and columns of the Data frame

```python
ds1.shape
```

<H1> Part 2 of the Assignment

<H2> Importing geographical coordinates of each postal code

```python
df1 = pd.read_csv('http://cocl.us/Geospatial_data')
df1
```

<H2> Merging the two Dataframes together

```python
Grouped_data = pd.merge(ds1,
                 df1[['Postal Code','Latitude','Longitude']],
                 on='Postal Code')
Grouped_data.head(10)
```

```python
Grouped_data['Borough'] = 'Toronto'
Grouped_data.head(10)
```

<H3> K means clustering and

```python

# set number of clusters
kclusters = 5

manhattan_grouped_clustering = Grouped_data.drop(['Neighbourhood','Borough','Postal Code'], 1)

manhattan_grouped_clustering 
# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(manhattan_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

```python
address = "Toronto, Canada"
geolocator = Nominatim(user_agent="toronto_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi in zip(Grouped_data['Latitude'], Grouped_data['Longitude'], Grouped_data['Neighbourhood']):
    label = folium.Popup(str(poi) + ' Cluster ', parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
