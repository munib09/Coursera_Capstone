<H2> Importing webpage into dataframe using Pandas

```python
import pandas as pd


dataframe = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')

df = dataframe[0]
df
```


<H2> Removing Values that are "Not Assigned" from the column Borough

```python
df.set_index('Borough', inplace = True)
dff= df.drop('Not assigned',axis = 0)
dff
```

<H2>Grouping Values in Neighbourhood column seperated by commas

```python
ds = dff.groupby(['Postcode','Borough'], as_index=True).agg(','.join)
ds.reset_index(inplace = True) 
ds
```

```python

ds.loc[ds.Neighbourhood == 'Not assigned', 'Neighbourhood'] = ds['Borough']
ds1= ds.rename(columns={'Postcode':'Postal Code'})
ds1
```
